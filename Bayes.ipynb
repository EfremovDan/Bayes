Возьмем 2 независимые переменные, которые распределены следующим образом: x_1 - нормальное распределение с матожиданием 2 и дисперсией 5, x_2 - равномерное распределение от 2 до 5. Случайная ошибка распределена логистически с нулевым матожиданием и дисперсией 1.

```python
import numpy as np
n = 1000
np.random.seed(123)
x_1 = np.random.normal(2, 5, n)
x_2 = np.random.uniform(2, 5, n)
u = np.random.logistic(0, 1, n)
```

Коэффициенты регресси возьмем как 1, 2 и 3.

```python
b_0, b_1, b_2 = 1, 2, 3
```

```python
y = b_0 + b_1*x_1 + b_2*x_2 + u
```

```python
import pystan
```

Проверим работу модели на одной итерации.

```python
%%time

data = {'n': n, 'x_1': x_1, 'x_2': x_2, 'y': y} 

model = pystan.stan(file = '/Users/efremovdanil/Desktop/Bayes/regression_normal.stan', data = data, chains = 1,
                    iter = 2000, seed=123)
```

```python
model
```

Значения получились похожими на истинные.

```python
from tqdm.notebook import tqdm
beta0 = [] #списки, в которые будем складывать средние значения коэффициентов, полученных моделью
beta1 = []
beta2 = []
for i in tqdm(range(100)):
    n = 1000
    x_1 = np.random.normal(2, 5, n) #также каждый раз заново генерируем новые рандомные данные
    x_2 = np.random.uniform(2, 5, n)
    u = np.random.logistic(0, 1, n)
    b_0, b_1, b_2 = 1, 2, 3
    y = b_0 + b_1*x_1 + b_2*x_2 + u
    
    data = {'n': n, 'x_1': x_1, 'x_2': x_2, 'y': y} 

    model = pystan.stan(file = '/Users/efremovdanil/Desktop/Bayes/regression_normal.stan', data = data, chains = 1,
                        iter = 2000, seed=123)
    
    beta0.append(model['beta_0'].mean())
    beta1.append(model['beta_1'].mean())
    beta2.append(model['beta_2'].mean())
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_percentage_error as mape
normal = []            # собираем в список 6 значений ошибок, первые 3 из которых - это RMSE по всем коэффициентам, 
                       # последние 3 - MAPE
print("RMSE for beta_0:", mse(beta0, np.ones(100) * 1, squared=False))
normal.append(mse(beta0, np.ones(100) * 1))
print("RMSE for beta_1:", mse(beta1, np.ones(100) * 2, squared=False))
normal.append(mse(beta1, np.ones(100) * 2))
print("RMSE for beta_2:", mse(beta2, np.ones(100) * 3, squared=False))
normal.append(mse(beta2, np.ones(100) * 3))
print()
print("MAPE for beta_0:", mape(beta0, np.ones(100) * 1))
normal.append(mape(beta0, np.ones(100) * 1) * 100)
print("MAPE for beta_1:", mape(beta1, np.ones(100) * 2))
normal.append(mape(beta1, np.ones(100) * 2) * 100)
print("MAPE for beta_2:", mape(beta2, np.ones(100) * 3))
normal.append(mape(beta2, np.ones(100) * 3) * 100)
```
